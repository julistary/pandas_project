### Importing libraries

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math
import src.visualization as v
```

### Loading datasets

```python
df_1 = pd.read_csv("dataframes/first.csv")
df_2 = pd.read_csv("dataframes/second.csv")
df_3 = pd.read_csv("dataframes/third.csv")
df_4 = pd.read_csv("dataframes/forth.csv")
df_5 = pd.read_csv("dataframes/fifth.csv")
df_6 = pd.read_csv("dataframes/sixth.csv")
```

## 1st hypothesis: The country with more attacks is Australia

```python
df_1.head()
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
attack_countries = sns.countplot(x=df_1.Country, palette="pastel")
ax.set_title("Attacks and countries")
```

```python
attack_countries.figure.savefig("images/1sth.svg")
```

## 2nd hypothesis: The region with more attacks is "Latin America and the Caribbean Sea"

```python
fig, ax = plt.subplots(figsize=(20, 8))
attack_regions = sns.countplot(x=df_2.Region, palette="pastel")
ax.set_title("Attacks and regions")
plt.xticks(rotation=30)
```

```python
attack_regions.figure.savefig("images/2ndh.svg")
```

## 3rd Hypothesis: The most frequent activity is surfing

```python
fig, ax = plt.subplots(figsize=(25, 8))
activity = sns.countplot(x=df_3.Activity_REGEX, palette="pastel")
ax.set_title("Attacks and activities")
```

```python
activity.figure.savefig("images/3rdh.svg")
```

## 4th hypothesis: The older the age, the more fatal attacks 

```python
fig, ax = plt.subplots(figsize=(15, 8))
plot = sns.histplot(x=df_4.Age, binwidth=10, hue= df_4.Fatal, palette = "pastel")
plt.legend(loc='upper right', labels=['Fatal', 'Not Fatal'])
plot.axvline(x=df_4.Age.mean(), c="violet", label="mean")
ax.set_title("Fatal attacks and age")
```

```python
fatal = sns.countplot(x=df_4.Fatal, palette="pastel")
```

```python
math.ceil(df_4.Age.mean())
```

```python
plot.figure.savefig("images/4thh.svg")
```

```python
fatal.figure.savefig("images/4_1thh.svg")
```

## 5th hypothesis: The trend of attacks tends to increase since 1975

```python
fig, ax = plt.subplots(figsize=(15, 8))
years = sns.lineplot(x=df_5["years"], y=df_5["number_of_attacks"])
ax.set_title("Attacks per year")
```

```python
years.figure.savefig("images/5thh.svg")
```

## 6th hypothesis: The shark that attacks the most is the white shark

```python
fig, ax = plt.subplots(figsize=(25, 8))
species = sns.countplot(x=df_6.Species_REGEX, palette="pastel")
ax.set_title("Attacks by Species")
```

```python
species.figure.savefig("images/6thh.svg")
```
